<h2>الگوریتم k-means</h2>
<div>
یکی از الگوریتم‌های ساده و نسبتاً پرکاربرد برای خوشه‌بندی الگوریتم k-means است. هدف ما در این تمرین پیاده‌سازی این الگوریتم نیست و فقط آَشنایی با آن است. برای استفاده از این الگوریتم و الگوریتم‌های دیگر خوشه‌بندی می‌توانید ماژول‌های <code>sklearn.cluster</code> را مشاهده کنید.</div>

```python
# your code here
labels = kmeans.fit_predict(X=X, y=y)
labels
```

<h2>تعیین تعداد خوشه‌ها</h2>
یکی از ضعف‌های الگوریتم k-means مشخص کردن تعداد خوشه‌هاست. در این مثال با مشاهده شکل توانسیم که تعداد خوشه‌ها را حدس بزنیم. اما در دنیای واقعی این کار قابل انجام نیست. یکی از راه‌ها کاهش بعد با استفاده از PCA است که در تمرین اول مشاهده کردید. روش دیگر استفاده از روش آرنج (elbow) است.


<font color='Green'> پرسش: اگر تعداد خوشه‌ها را برای داده‌های قبل 2 در نظر بگیریم، نمودار قبلی را برای این حالت رسم کنید. دقت کنید که <code>random_state=42</code> باشد. </font>

```python
# your code here
kmeans_k2 = KMeans(n_clusters=2, random_state=42)
labels_k2 = kmeans_k2.fit_predict(X=X, y=y)
print(labels_k2)
plt.scatter(X[:, 0], X[:, 1], cmap=plt.cm.cool, c=labels_k2)
```

<div dir=rtl>
<h2>روش آرنج (elbow)</h2>
شما در درس با روش آرنج آشنا شده‌اید.

<font color='Green'> پرسش: روش آرنج را برای مقدار خوشه‌های 1 تا 10 حساب کنید و نمودار میزان تابع هزینه را برای هر خوشه رسم کنید. برای هر مرحله <code>random_state</code> را برابر با صفر در نظر بگیرید.  
</div>

```python
# your code here
costs = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    result = kmeans.fit(X)
    costs.append(result.inertia_)

plt.plot(range(1,11), costs, "o-")
```

<div dir=rtl>
<h2>خوشه‌بندی‌های پیچیده‌تر</h2>
همیشه خوشه‌بندی‌ها به این صورت نیست. همچنین k-means تنها الگوریتم خوشه‌بندی نیست. در زیر توزیع‌های مختلف داده‌های مختلف با روش k-means رسم شده‌است.

 شما می‌توانید مشاهده کنید که در کدام حالت این الگوریتم خوب عمل می‌کند.

تمامی این‌ کدها از سند مطالب <code>sklearn</code> استفاده شده است. شما می‌توانید این اسناد را در <a href="https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_assumptions.html"> اینجا</a> مشاهده کنید.


همچنین می‌توانید الگوریتم‌های دیگر را نیز در اسناد <code>sklearn</code> مشاهده کنید.
</div>

```python
sns.set_style('white')

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
plt.title("Anisotropicly Distributed Blobs")

# Different variance
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Unequal Variance")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Unevenly Sized Blobs")

plt.show()
```

این الگوریتم نیز مانند PCA کار کاهش بعد را انجام میدهد. با استفاده از این الگوریتم (موجود در <code>sklean.manifold.Isomap</code>) داده‌ها را به داده‌هایی با بعد 2 تبدیل کنید. پس از آن دو نمودار رسم کنید. نمودار اول رنگ هر داده را با لیبل‌هایی که الگوریتم kmeans تولید کرده است تعیین کنید و نمودار دوم رنگ هر داده برابر با digits.target که همان لیبل کلاس است، فرض بگیرید. چه چیزی متوجه می‌شوید؟

```python
from sklearn.datasets import load_digits
digits = load_digits()

# your code here
kmeans = KMeans(n_clusters = 10, random_state = 42)
X = digits.data
kmeans.fit_predict(X)
centers=list(map(lambda x: x.reshape(8,8),kmeans.cluster_centers_))
fig, axes=plt.subplots(ncols=10)
for i,c in enumerate(centers):
  axes[i].imshow(c)
```

### در پاسخ به پرسش اول، با توجه به خروجی ای که دریافت کرده ایم، می‌توانیم ببینیم که با این کار رزولوشن و کیفیت تصویر کاهش می یابد اما همچنان می‌توانیم به طور حدودی و با تشخیص مرکز هر دسته برچسب آن دسته را نیز تشخیص دهیم.

### استفاده از Isomap

```python
from sklearn.manifold import Isomap

embedding = Isomap(n_components=2)
transformed_X = embedding.fit_transform(X)
labels = kmeans.fit_predict(X)
fig,axes=plt.subplots(ncols=2,figsize=(20,10))
axes[0].scatter(transformed_X[:, 0], transformed_X[:, 1], cmap=plt.cm.cool, c=labels)
axes[1].scatter(transformed_X[:, 0], transformed_X[:, 1], cmap=plt.cm.cool, c=digits.target)
axes[1].set_ylabel("true label")
axes[0].set_ylabel("predicted label")
```

<h1>تعیین اپسیلون</h1>

<font color='Green'> پرسش: علت استفاده از KNN برای پیدا کردن اپسیلون بهینه را شرح دهید. سپس با رسم گراف k-distance  و با پیروی از الگوی آرنج مقدار اپسیلون بهینه را   بدست آورید.

<font color='Green'>  برای رسم گراف k-distace ‌می‌توانید از ماژول NearestNeighbors در کنابخانه Scikit-learn و برای مشخص کردن مقدار اپسیلون از کتابخانه kneed استفاده کنید.

### برای اینکه مقدار اپسیلون را تعیین کنیم باید بتوانیم مقدار حدودی چگالی خوشه ها را بدست آوریم یا از قبل بدانیم. هر چه فاصله نقاط بیشتر باشد و چگالی کمتر باشد، باید اپسیلون را بزرگتر در نظر بگیریم تا بتوانیم تمامی نقاط را پوشش داده و به یک دسته مد نظر اضافه کنیم. الگوریتم کی‌ان‌ان به این شکل عمل می‌کند که فاصله هر نقطه را تا ان نقطه اطرافش بدست آورده و با کمک روشی مشخص، عددی که به نظر می رسد حداکثر فاصله یک نقطه برای آنکه به یک خوشه تعلق داشته باشد است را بدست می آورد و اگر نقطه ای فاصله اش کمتر از این عدد باشد پس به آن خوشه تعلق دارد و در غیر این صورت ندارد.

```python
!pip install --upgrade kneed

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import v_measure_score
from kneed import KneeLocator
import pandas as pd 
```

```python
# your code here 
neighbors = NearestNeighbors(n_neighbors=2).fit(X_aniso)
ds, indices = neighbors.kneighbors(X_aniso)
ds = np.sort(ds[:,1], axis=0)
kl = KneeLocator(range(len(ds)), ds,curve="convex", direction="increasing")
kl.plot_knee()
plt.show()
aniso_eps = kl.knee_y
```

```python
neighbors = NearestNeighbors(n_neighbors=2).fit(X_varied)
ds, indices = neighbors.kneighbors(X_varied)
ds = np.sort(ds, axis=0)[:,1]
kl = KneeLocator(range(len(ds)), ds,curve="convex", direction="increasing")
kl.plot_knee()
plt.show()
varied_eps = kl.knee_y
```


<div dir=rtl >
<h1>تعیین MinPts</h1>

در مرحله قبل برای تعیین اپسیلون بهینه  مقدار MinPts را نیز مشخص کردیم. اما لزوما با آن مقدار  بهترین خوشه‌بندی را نخواهیم داشت.

<font color='Green'> پرسش: در این مرحله با در نظر گرفتن اپسیلون بدست آمده و با آزمودن مقادیر مختلف برای  پارامتر MinPts، بهترین حالت برای خوشه‌بندی را با استفاده از رنگ‌بندی مناسب رسم کنید. 

</div>




```python
# your code here
# x_aniso
fig,axes= plt.subplots(ncols=4,nrows=3,figsize=(20,20))
for i in range(12):
   db = DBSCAN(eps= aniso_eps, min_samples=i+1)
   labels=db.fit_predict(X_aniso)
   axes[i//4][i%4].scatter(X_aniso[:,0], X_aniso[:,1], c=labels)
   axes[i//4][i%4].set_title(f"minpts = {i+1}, classes={len(set(labels))}")
```

```python
# x_varied
fig,axes= plt.subplots(ncols=4,nrows=3,figsize=(20,20))
for i in range(12):
   dbscan = DBSCAN(eps= varied_eps, min_samples=i+1)
   labels=dbscan.fit_predict(X_varied)
   axes[i//4][i%4].scatter(X_varied[:,0], X_varied[:,1], c=labels)
   axes[i//4][i%4].set_title(f"minpts = {i+1}, classes={len(set(labels))}")
```

### بهترین مقدار برای انیسو برابر با شش و برای ورید برابر با ۹ است.
